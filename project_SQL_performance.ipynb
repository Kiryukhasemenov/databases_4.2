```python
def phone_formatting(phone):
    if phone.startswith('+'):
        phone = member_phone[1:]
    phone = re.sub(r'[\(\) \-]', r'', phone)
    
    if phone =='':
        phone = None

    return phone
    
def isactive_formatting(isactive):
    isactive = isactive.lower()
    if isactive in ['да', '1', 'yes', '+', 'y']:
        isactive = 1
    else:
        isactive = 0
    return isactive

def curr_sit_formatting(curr_sit):
    sit_dict = {0: 'not started', 1:'in process', 2:'done'}
    curr_sit = sit_dict[curr_sit]
    return curr_sit
        
```

### Изменение существующих данных

```python
def upgrading():
    table = input('Введите название таблицы: ')
    if table == 'members':
        
        print('Вы можете изменить активность участника, его контакты и должность.')
        member_name = input('Введите имя участника: ')
        #to_upgrade = input('Нажмите 1 для изменения активности участника, 2 - для изменения контактов, 3 - для изменения должности')
        #change_dict = {}
        change_isactive = input('Нажмите 1 для изменения активности участника, или Enter, если менять ее не надо')
        if change_isactive == '1':
            new_isactive = input('теперь участник активен?')
            new_isactive = isactive_formatting(new_isactive)
            
            #change_dict['member_isactive'] = new_isactive
            cursor.execute("""UPDATE members SET member_isactive = %s WHERE member_name = %s""", (new_isactive, member_name))
            db.commit()
            
        change_contacts = input('Нажмите 1 для изменения контактов участника, или Enter, если менять их не надо')
        if change_contacts == '1':
            change_phone = input('Поменялся ли телефон? Нажмите 1, если да, и Enter, если нет')
            if change_phone == '1':
                new_phone = input('Введите новый номер телефона: ')
                new_phone = phone_formatting(new_phone)
                #change_dict['member_phone'] = new_phone
                cursor.execute("""UPDATE members SET member_phone = %s WHERE member_name = %s""", (new_phone, member_name))
                db.commit()
                
            change_email = input('Поменялся ли емейл? Нажмите 1, если да, и Enter, если нет')
            if change_email == '1':
                new_email = input("Введите новый емейл")
                #change_dict['member_email'] = new_email
                cursor.execute("""UPDATE members SET member_email = %s WHERE member_name = %s""", (new_email, member_name))
                db.commit() 
        change_position = input('Нажмите 1 для изменения должности участника, или Enter, если менять ее не надо')
        if change_position == '1':
            new_position = input('Введите новую должность участника: ')
            #change_dict['member_position'] = new_position
            cursor.execute("""UPDATE members SET member_position = %s WHERE member_name = %s""", (new_position, member_name))
    
            db.commit()
        
    elif table == 'texts':
        print('Вы можете изменить нынешний формат и статус текста.')
        text_name = input('Введите название текста: ')
        new_format = input('Введите новый формат текста: ')
        
        
        cursor.execute('''SELECT DISTINCT text_situation FROM texts''')
        statuses_ = cursor.fetchall()
        statuses_ = [s[0] for s in statuses_]
        print('Статусы бывают следующие: ')
        for s in statuses_:
            print(s)
        new_status = input('Введите новый статус текста: ')
        
        cursor.execute("""UPDATE texts SET text_type = %s, text_situation = %s WHERE text_name = %s""", (new_format, new_status, text_name))
        
        db.commit()#member_query = """UPDATE users SET name = 'Kareem' WHERE id = 1"""
    else:
        print('Вы не можете редактировать эту таблицу или такой таблицы нет вовсе.')
    
    #db.commit()
    return 'Изменения записаны. Если вы хотите заменить еще что-то - запустите программу снова.'
```
